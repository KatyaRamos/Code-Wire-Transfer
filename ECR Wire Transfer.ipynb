```python
import pandas as pd
from datetime import date

ewt1 = pd.read_excel(r'C:\Users\Katya Ramos-Barrios\Documents\GitHub - Work Samples\Fake Data\Fake Wire Transfers.xlsx')
ewt1
```

```python
country_reftab = pd.read_excel(r'C:\Users\Katya Ramos-Barrios\Documents\GitHub - Work Samples\Fake Data\ISO_Countryname_codelist.xlsx')
country_reftab.head()
```

```python
ewt2=pd.merge(ewt1[['HUB_TXN_ID', 'PRE_FE_ORIG_COUNTRY_ID']],country_reftab,left_on=['PRE_FE_ORIG_COUNTRY_ID'],right_on=['COUNTRY_ISO']).drop(['COUNTRY_ISO'],1)
ewt2 = ewt2.rename({'NAME': 'PRE_FE_ORIG_COUNTRY_NAME'}, axis=1)
ewt2
```

```python
ewt3=pd.merge(ewt1[['HUB_TXN_ID', 'PRE_FE_RECI_COUNTRY_ID']],country_reftab,left_on=['PRE_FE_RECI_COUNTRY_ID'],right_on=['COUNTRY_ISO']).drop(['COUNTRY_ISO'],1)
ewt3 = ewt3.rename({'NAME': 'PRE_FE_RECI_COUNTRY_NAME'}, axis=1)
ewt3
```

```python
ewt1_2=pd.merge(ewt2,ewt3,on='HUB_TXN_ID', how='left')
ewt1_2
```

```python
ewt=pd.merge(ewt1,ewt1_2[['HUB_TXN_ID','PRE_FE_ORIG_COUNTRY_NAME','PRE_FE_RECI_COUNTRY_NAME']],on='HUB_TXN_ID', how='left' )
ewt
```

```python
ewt.isnull().sum()
```

```python
ewt['DATE'] = ewt['RAW_TAG32'].map(str).map(lambda x: "20{}-{}-{}".format(x[5:7],x[7:9],x[9:11]))
ewt['CURRENCY'] = ewt['RAW_TAG32'].map(lambda x: x[11:14])
ewt['TRANSACTION_VALUE'] = ewt['RAW_TAG32'].map(lambda x: x[14:]).str.replace(",",".").astype(float)
ewt.head()
```

```python
#Convert date column as object to date
ewt['DATE'] = pd.to_datetime(ewt['DATE'])
ewt.head()
```

## ECR Domestic SWIFT Wire Transfer

```python
ewt['DOMESTIC_TRANSACTION'] = (ewt['PRE_FE_ORIG_COUNTRY_ID']==ewt['PRE_FE_RECI_COUNTRY_ID'])
ewt1_DOM = (ewt['DOMESTIC_TRANSACTION']==True)
ewt1_DOMT=ewt[ewt1_DOM]
ewt1_DOMT.head()
```

### ECR Domestic SWIFT Number of Clients Wire Transfer

```python
# of Clients - SWIFT Wire Transfer (Unique id count FOR COMPLETE RISK ASSESSMENT PERIOD)
ECR_DWOTC=pd.pivot_table(ewt1_DOMT, values='ORIG_ENTITY', index=['PRE_FE_ORIG_COUNTRY_NAME','PRE_FE_ORIG_BANK_NAME','PRE_FE_RECI_COUNTRY_NAME', 'PRE_FE_RECI_BANK_NAME'], aggfunc= pd.Series.nunique , margins = True)
ECR_DWOTC
```

### ECR Domestic SWIFT Transaction Volume Wire Transfer

```python
# Total Transactions
ECR_DWOTTT=pd.pivot_table(ewt1_DOMT, values='HUB_TXN_ID', index=['PRE_FE_ORIG_COUNTRY_NAME','PRE_FE_ORIG_BANK_NAME','PRE_FE_RECI_COUNTRY_NAME', 'PRE_FE_RECI_BANK_NAME'], aggfunc= 'count' , margins = True)
ECR_DWOTTT
```

### ECR Domestic SWIFT Total Transaction Value Wire Transfer

```python
#NEEDS CURRENCY EXCHANGE TO SUM ALL TRANSACTIONS IN CAD
ECR_DWTV=pd.pivot_table(ewt1_DOMT, values='TRANSACTION_VALUE', index=['PRE_FE_ORIG_COUNTRY_NAME','PRE_FE_ORIG_BANK_NAME','PRE_FE_RECI_COUNTRY_NAME','PRE_FE_RECI_BANK_NAME','CURRENCY'], aggfunc= 'sum' , margins = True)
ECR_DWTV
```

## ECR International SWIFT Wire Transfer

```python
ewt1_INT = (ewt['DOMESTIC_TRANSACTION']==False)
ewt1_INTT=ewt[ewt1_INT]
ewt1_INTT.head()
```

### ECR International SWIFT Total Customers Wire Transfer

```python
# Total Customer - SWIFT Wire Transfer (Unique id count FOR COMPLETE RISK ASSESSMENT PERIOD)
ECR_IWOTC=pd.pivot_table(ewt1_INTT, values='ORIG_ENTITY', index=['PRE_FE_ORIG_COUNTRY_NAME','PRE_FE_ORIG_BANK_NAME','PRE_FE_RECI_COUNTRY_NAME','PRE_FE_RECI_BANK_NAME'], aggfunc= pd.Series.nunique , margins = True)
ECR_IWOTC
```

### ECR International SWIFT Transaction Volume Wire Transfer

```python
ECR_IWOTTT=pd.pivot_table(ewt1_INTT, values='HUB_TXN_ID', index=['PRE_FE_ORIG_COUNTRY_NAME','PRE_FE_ORIG_BANK_NAME','PRE_FE_RECI_COUNTRY_NAME','PRE_FE_RECI_BANK_NAME'], aggfunc= 'count' , margins = True)
ECR_IWOTTT
```

### ECR International SWIFT Total Transaction Value Wire Transfer

```python
#NEEDS CURRENCY EXCHANGE TO SUM ALL TRANSACTIONS IN CAD
ECR_IWTV=pd.pivot_table(ewt1_INTT, values='TRANSACTION_VALUE', index=['PRE_FE_ORIG_COUNTRY_NAME','PRE_FE_ORIG_BANK_NAME','PRE_FE_RECI_COUNTRY_NAME','PRE_FE_RECI_BANK_NAME','CURRENCY'], aggfunc= 'sum' , margins = True)
ECR_IWTV
```
